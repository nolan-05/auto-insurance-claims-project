# Preliminary Poisson Frequency Model

## This section will create a preliminary poisson claim frequency model. Claim count is going to be used as the response variable, and exposure will also be included as policies are observed for different lengths of time. *This model is only a baseline frequency model and not a full pricing model*. 

## Model Setup

##### ClaimNb = response variable
##### Exposure = offset / exposure adjustment 

##### Predictors: VehicleAgeBand, DriverAgeBand, BonusMalusBand, VehBrand, VehGas, PopulationDensityBand

## Fitting the Poisson Model

#### Importing python packages

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

#### Creating Poisson model

```python
poisson_model = smf.glm(
    formula='ClaimNb ~ C(VehicleAgeBand) + C(DriverAgeBand) + C(BonusMalusBand, Treatment(reference="50")) + C(VehBrand) + C(VehGas) + C(PopulationDensityBand, Treatment(reference="Low"))',
    data=df,
    family=sm.families.Poisson(),
    offset=np.log(df["Exposure"])
).fit()
```

### Interpreting the Model

##### This model uses claim count as the response variable, and the log of exposure as an offset. The model is estimating claim frequency based on different policy exposure lengths. All variables are interpreted relative to their baseline categories, so coefficients should be read as relative changes to the baseline level for each category. 
##### The following variables were excluded from the model with an explanation next to them: 
##### Area and Region: Area, Region and PopulationDensityBand overlap very closely as they're all geographic location variables. This caused many large coefficients which were unstable. In order to fix the model and maintain stability PopulationDensityBand was chosen since it is a variable which is directly measurable. 
##### VehPower: In the analysis it did not show a clear association with expected claim frequency, and upon adding it to the model it did not make much, if any, difference on other variables. To prevent overcomplication and potential false associations, it has been excluded. 

#### The baselines for each category are as follows: 
###### Area: A 
###### DriverAgeBand: 16-24
###### VehicleAgeBand: 0-4
###### BonusMalusBand: 50 
###### VehBrand: B1
###### VehGas: Diesel 
###### PopulationDensityBand: Extremely High

#### Below we can view the coefficients of each categorical variable compared to our baselines.

```python
model_coefficients = np.exp(poisson_model.params)
model_coefficients
```

 #### Area coefficients interpretation

##### Area A is the reference category. The area values below are exponentiated coefficients, so they can be interpreted as claim frequency relativities compared with Area A, while holding the other model variables constant. 
##### Area B has a relativity of 1.043782, which reflects a minimal increase in claim frequency from our baseline of Area A, holding all other variables constant. On the other hand, Area E has a very large relativity of 3.026643. This reflects a 3x increase from A in claim frequency, holding all else constant. Other relativities are also different from baseline, with Area D's at 2.942593, F at 1.706520, and C at 1.349360.
##### Overall findings suggest that area is meaningfully associated with claim frequency.
##### One possible explanation for this association is that different geographical areas have different traffic flows. More intense traffic flows could plausibly cause more accidents, even if drivers are experienced. 

#### DriverAgeBand coefficients interpretation

```python
DriverAgeBand_coefficients = np.exp(poisson_model.params).filter(like="DriverAgeBand")
DriverAgeBand_coefficients
```

##### DriverAgeBand used 16-24 as the baseline age range, thus each coefficient will be claim frequency compared to claim frequency of 16-24 year olds, *all else held constant*. 
##### Drivers aged 25-34 had a relativity of 0.675846 and drivers aged 35-44 had a relativity of 0.860656. Both of these mean that compared to drivers aged 16-24, they each have far lower claim frequency. The exact interpretation is that drivers aged 25-34 have ~67.5846% of the claim frequency that drivers aged 16-24 do. The age bands of 45-54 and 65+ have a slightly higher relativity than baseline, and the age band of 55-64 has a slightly lower relativity than baseline.
##### Overall findings suggest that driver age is meaningfully associated with claim frequency.
##### One possible explanation for this association could be that younger drivers are less experienced, so more accidents, while middle aged drivers have good reaction time and experience, while older drivers have weaker reaction times.

#### VehicleAgeBand coefficients interpretation

```python
VehicleAgeBand_coefficients = np.exp(poisson_model.params).filter(like="VehicleAgeBand")
VehicleAgeBand_coefficients
```

##### VehicleAgeBand used 0-4 as the baseline, which means that *all else held constant*, each age band was compared to the claim frequency of a car aged 0-4 years old. 
##### The relativities of vehicle ages appear to decrease from baseline the older a vehicle is. 5-9 has a relativity of 0.884249, which is the smallest change from baseline. The other relativities were 0.790362 for 10-14, 0.606928 for 15-19, and 0.527036 for 20+ which was the largest change from baseline. 
##### Findings so far suggest that there is a meaningful association between vehicle age and claim frequency.
##### One possible explanation for this association is the driving and repair patterns of older cars versus newer cars, though the information in the dataset is far too limited to find the exact cause. 

#### BonusMalusBand coefficients interpretation

## Limitations

##### There are several limitations of this model. The first, and most trivial to mention, is that this is a preliminary claim frequency model *only*, and is not a claim severity or even a pricing model. This model predicts expected claim frequency per policy year, nothing more. 
##### The Poisson distribution (used for modelling) assumes that mean ≈ variance, which is not always true for insurance claim data. 
##### Banding variables was able to improve interpretation, however in the process it also reduced detail in the predictions. Tighter bands, or removing them altogether, could reduce the severity of this limitation, but interpretability is also important for a model.
##### Area and region, which could be two large predictors, had to be left out to avoid geographic area overlap and unstable predictions. 
##### This model shows *associations*, no *causation* has been proven with this model. Any causation is purely speculative. 
##### In the future, better data analysis, severity and pricing modelling, overdispersion checks, and validation are all good areas to improve upon. 

## Assumptions

##### 1. Claim counts follow a Poisson distribution.

##### 3. The mean claim count is directly proportional to exposure, which is why log(Exposure) was used as an offset.
##### 4. The logarithm of expected claim frequency is linear in the predictors.
